## WoE

Cуществует показатель, позволяющий сравнивать влияние признака на таргет - **WoE** (Weight of evidence)  
$$WoE_{x_{i}} = Ln\left(\frac{B_{i}}{G_{i}}\right) - Ln\left(\frac{B}{G}\right) = Ln\left( \frac{bad\_rate_{i}}{1 - bad\_rate_{i}}\right) - Ln\left(\frac{bad\_rate}{1 - bad\_rate}\right)$$   
Где $B_{i}$ - количество просрочек при значении признака $X = x_{i}$,  
$G_{i}$ - количество заявок без просрочки при значении признака $X = x_{i}$,  
$B$ - общее число просрочек  
$G$ - общее число непросроченных заявок  

Смысл - отличие логарифма шанса при данном значении признака от логарифма среднего шанса.  
Для бинарного признака находятся два значения, для дискретного - для каждого значения признака свое значения woe,  
а для непрерывного или дискретного с большим количеством значений считаем woe по бакетам  
(на которые предварительно группируем признак). 

Вспомогательная функция разбивает признак на бакеты и считает на них bad_rate,  
также дополнительно: сумму просрочек в бакете, среднее значение предсказания, количество  
заявок в бакете и среднее значения признака.

```python
def bad_rate(df,feature,target,num_buck = 10):
    '''
    df - датафрейм
    feature - признак (название столбца в датафрейме)
    target - название целевой переменной
    num_buck - количество бакетов
    Функция считает bad_rate для каждого бакета
    Возвращает датафрейм с аггрегациями (сумма просрочек в бакете, среднее значение предсказания,
    количество заявок в бакете, среднее значение признака в бакете)
    '''
    return df.assign(bucket = np.ceil(df[feature].rank(pct = True) * num_buck),obj_count = 1)\
             .groupby('bucket')\
             .agg({target:'sum','obj_count':sum,feature:'mean'})\
             .rename(columns = {target:'target_sum',feature:'average'})\
             .assign(bad_rate = lambda x:x.target_sum/x.obj_count)
```

```python
from math import log
def woe(df,feature,target,num_buck = 10):
    '''считаем woe
       df - датафрейм
       feature - признак
       predict - предсказание
       target - целевая переменная
       num_buck - количество бакетов
       Считаем woe, так как признаки могут принимать значения 0, добавляем малую константу 0.00001
    '''
    agg = bad_rate(df,feature,target,num_buck).reset_index()
    agg = agg[agg.target_sum != 0]
    return agg.assign(woe = lambda x:(x.bad_rate/(1-x.bad_rate) + 0.00001).apply(log) -
                      log((df[target].sum()/(len(df) - df[target].sum())))).set_index('bucket')
```

## Линейность WoE

На первом этапе корректного построения логистической регрессии необходимо проверить  
линейность признаков по WoE, линейность признаков позволяет сравнивать веса признаков между собой,  
а также позволяет получить более стабильный результат.  
**Проверка**: Обычно совмещают построение графика совместного распределения признака и woe,  
с подсчетом r2_score простой линейной регрессии.  
**Борьба**: Также как и в линейной регрессии либо линеаризуют монотонным преобразованием,  
либо разбивают нелинейную переменную на несколько линейных.

```python
#Взята из семинара2
from sklearn.linear_model import LinearRegression

def simple_reg(predictor, target):
    '''
    predictor - массив значений предиктора
    target - массив значений таргета
    Строим простую линейную регрессию 
    Будем использовать для отрисовки прямой и подсчета r2'''
    check = LinearRegression()
    check.fit(predictor.reshape(-1,1),target)
    return hv.Curve((np.array([i for i in np.linspace(min(predictor),max(predictor),100)]),
                     check.coef_*np.array([i for i in np.linspace(min(predictor),
                                                                  max(predictor),100)]) + check.intercept_))
```

```python
def paint_curve(df,feature,target,num_bucks = 10):
    '''
    df - датафрейм 
    feature - признак по которому проверяем линейность
    target - целевая переменная
    num_buck - число бакетов
    считаем woe, и по полученным точках строим лучшую прямую (с наименьшим r_2)
    '''
    return woe(df,feature,target,num_bucks).pipe(lambda x:hv.Scatter(zip(np.array(x.average), np.array(x.woe)),
                               kdims = ['average'] ,vdims=['woe'],
                               label = f'Зависимость woe от {feature}')* 
                                                 simple_reg(np.array(x.average),np.array(x.woe)))
```

### Проверка с помощью r2_score

Для признака строим простую линейную регрессию для бакетов получаемых из функции `woe`  
и считаем `r2_score` полученного результата  

```python
from sklearn.metrics import r2_score

def r_2check(df,feature,target,num_bucks = 10):
    '''
    df - датафрейм
    feature - название признака
    target - целевая переменная
    transform - преобразование над признаком
    Функция вычисляет r2 признака, она работает в двух режимах,
    если transform = None, то возвращается только r2 признака,
    если transform непуст, то возвращается r2 признака до преобразования и после
    '''
    m = LinearRegression()
    m.fit(np.array(woe(df,feature,target,num_bucks).average).reshape(-1,1),
          np.array(woe(df,feature,target,num_bucks).woe))

    return r2_score( np.array(woe(df,feature,target,num_bucks).woe),
                    m.predict(np.array(woe(df,feature,target,num_bucks).average).reshape(-1,1)))    
```

###  Проверка стабильности по времени

Модель строится на старых данных, так как просрочки появляются далеко не сразу, возникает необходимость  
убедиться, что на новых данных модель будет работать также, для этого нужно убедиться, что  
распределение не меняется со временем и woe стабильно по времени.  
Первое проверяется построением распредления признаки по кварталам,  
второе - построением графика woe по кварталам.  
Данная проверка выполняется на всей выборке.


```python
def simple_logreg(feature,target):
    '''
    feature - признак (стобец датафрейма)
    target - целевая переменная
    Простая логистическая регрессия,
    возвращает вектор вероятностей дефолта, 
    используется для построения графика стабильности по времени 
    '''
    model = make_pipeline(StandardScaler(),LogisticRegression())
    model.fit(np.array(feature).reshape(-1,1),np.array(target))
    return model.predict_proba(np.array(feature).reshape(-1,1))[:,1]
```

```python
import matplotlib.pyplot as plt
from matplotlib import rc
from itertools import product
from datetime import datetime

def distribution(df,feature,date,num_bucks = 10):
    '''
    df - датафрейм
    feature - название признака
    date - дата
    num_bucks - число бакетов
    Отрисовка распределения бакетов признака по кварталам
    '''
    agg =  df.assign(bucket = np.ceil(df[feature].rank(pct = True) * num_bucks),obj_count = 1)\
             .groupby([date,'bucket'])\
             .agg({'obj_count':sum,feature:'mean'})
    agg = agg.reset_index().assign(bucket = lambda x:x.bucket.astype(int)).set_index([date,'bucket'])
    index = agg.reset_index()[date].unique()
    groups = agg.reset_index()['bucket'].unique()
    keys = product(index,groups)
    bars = hv.Bars([k +((agg.query(f'{date} == "{k[0]}" and bucket == {k[1]}')
                         .reset_index().obj_count.values[0]) / 
                         (agg.query(f'{date} == "{k[0]}"').groupby([date])[['obj_count']].sum().values[0][0]),) for k in keys],
               ['date', 'bucket'], 'obj_count')
    return bars.relabel(group='Stacked')   
```

```python
def woe_stab(df, feature, date, target, num_buck = 5):
    '''
    df - датафрейм
    feature - название признака
    date - дата
    target - целевая переменная
    num_buck - число бакетов
    Отрисовка графика woe по кварталам
    '''
    df = df.assign(predict = simple_logreg(df[feature].astype(np.float64),df[target]))
    agg = df.assign(bucket = np.ceil(df[feature].rank(pct = True) * num_buck),obj_count = 1)\
            .groupby(['bucket',date])\
            .agg({target:'sum','predict':'mean','obj_count':sum,feature:'mean'})\
            .rename(columns = {target:'target_sum',feature:'av_f'})\
            .assign(bad_rate = lambda x:x.target_sum/x.obj_count)
    agg = agg.assign(nums = agg.groupby('quarter')['obj_count'].transform('sum'),
                 bad_nums = agg.groupby('quarter')['target_sum'].transform('sum')) 
    agg = agg.assign(woe = lambda x:((x.bad_rate/(1-x.bad_rate)) + 0.000001).apply(log) - 
             (x.bad_nums / (x.nums - x.bad_nums)).apply(log)).reset_index()
    agg = agg[agg.target_sum != 0]
    return hv.Overlay([hv.Curve(agg.query(f'{d}== bucket'), kdims=[date], vdims=['woe'],label=f'{d}')
             .redim.range(HL_auto=(0,50)).redim.label(HL_clear='Статистика Хосмер-Лемешева')
              for d in set(agg.bucket)])
```

## Information value 

Метрика позволяющая оценивать силу зависимости между признаком и целевой переменной (совокупную разделяющую силу признака),  
для решения вопроса о включение признака в модель, возможно отсортировать признаки по IV  
и добавить с наиболее высокими значениями
$$IV = \sum_{i=1}^{k}\left(\frac{B_{i}}{B} - \frac{G_{i}}{G}\right) * WoE_{x_{i}}$$



####  Задание 3 

 - Реализуйте IV (1 балл)
 - Отсортируйте признаки по IV и постройте график зависимости roc_auc от числа признаков, напишите выводы (на всех признаках) (2 балла)

```python
def IV(df, feature, target, num_buck = 10):
    '''df - датафрейм
       feature - название признака для которого считать IV
       num_buck - число бакетов
    '''
    Bad = sum(df[target] == 1)
    Good = sum(df[target] == 0)
    
    br = bad_rate(df, feature, target, num_buck)

    return br.assign(good_x = lambda x: x.obj_count - x.target_sum) \
             .assign(IV_per_bucket = lambda x: (x.target_sum / Bad - x.good_x / Good) * \
                     ((x.bad_rate / (1-x.bad_rate) + 0.00001).apply(log) -
                      log(df[target].sum() / Good)))['IV_per_bucket'].sum()
```

В данном коде отрисовывает зависимость roc_auc от количества признаков,  
признаки собираются случайным образом, вам нужно будет задать  
порядок по значению IV, начиная с самых важных признаков.
